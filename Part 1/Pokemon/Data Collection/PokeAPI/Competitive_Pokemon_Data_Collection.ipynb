# Worlds 2024

```python
url = 'https://victoryroadvgc.com/2024-worlds/'
```

```python
# Make a GET request to fetch the raw HTML content
html_request = requests.get(url)
html_request
```

```python
# Parse the html content
Tournament_Results = BeautifulSoup(html_request.content, "html.parser")
Tournament_Results
```

```python
#Pulling the Rankings from the website
Rankings = Tournament_Results.find_all('td', style = 'border: none;')
Rankings = [Rankings[i].get_text() for i in range(len(Rankings))]
#remove '/xa0' from the list
Rankings = [Rankings[i].replace('\xa0', '') for i in range(len(Rankings))]
#remove empty elements
Rankings = [Rankings[i] for i in range(len(Rankings)) if Rankings[i] != '']
#remove '/xa0' from the list
Rankings = [Rankings[i].replace('\xa0', '') for i in range(len(Rankings))]
#remove any elements containing $
Rankings = [Rankings[i] for i in range(len(Rankings)) if '$' not in Rankings[i]]


print(Rankings)
Rankings


```

```python
#remove everything after 'Anthony-Christopher Manolache' in the list ensure to keep the name
Rankings = Rankings[:Rankings.index('Anthony-Christopher Manolache(Astora)')]
#add the name back to the list
Rankings.append('Anthony-Christopher Manolache(Astora)')
print(Rankings)
```

```python
#if an element only contains a number, make it a new list called Rank
Rank = []
for i in range(len(Rankings)):
    if Rankings[i].isdigit():
        Rank.append(Rankings[i])
#add rank to a dataframe called worlds_2024_results
worlds_2024_results = pd.DataFrame(Rank, columns = ['Rank'])
worlds_2024_results
```

```python
#if an element contains a number and a - character change the - to a / character
record = [Rankings[i].replace('-', '-') for i in range(len(Rankings)) if '-' in Rankings[i]]
#remove the last element from the list
record = record[:-1]
#remove any element that contains a letter
record = [record[i] for i in range(len(record)) if not any(c.isalpha() for c in record[i])]
#add record to the dataframe

worlds_2024_results['Record'] = record
worlds_2024_results
```

```python
#take any element that contains a letter and add it to a new list called name
name = [Rankings[i] for i in range(len(Rankings)) if any(c.isalpha() for c in Rankings[i])]
len(name)
#add name to the dataframe
worlds_2024_results['Name'] = name
worlds_2024_results
```

```python
Teams = Tournament_Results.find_all('a', href = True)
#only keep elements that contain 'pokepast'
# convert the list so each element is a string
Teams = [str(Teams[i]) for i in range(len(Teams)) if 'pokepast' in Teams[i]['href']]
#remove all elements that don't contain 'pokepast'
Teams = [Teams[i] for i in range(len(Teams)) if 'pokepast' in Teams[i]]
#remove the bottom 8 elements
Teams = Teams[:-8]
#only keep characters between 9 and 30 in each element
Teams = [Teams[i][9:45] for i in range(len(Teams))]
Teams
```

```python
# Initialize an empty string to store combined HTML content
combined_html = ""

# Iterate over the list of URLs
for i in range(len(Teams)):
    url = Teams[i]
    html_request = requests.get(url)
    
    # Append the HTML content of each URL to the combined string
    combined_html += html_request.content.decode("utf-8")

# Parse the combined HTML content using BeautifulSoup
Team_Data = BeautifulSoup(combined_html, "html.parser")
```

```python
Worlds_2024_Teams = Team_Data
Worlds_2024_Teams = Worlds_2024_Teams.find_all('pre')
#strip the html tags from the list
Worlds_2024_Teams = [Worlds_2024_Teams[i].get_text() for i in range(len(Worlds_2024_Teams))]
#seperate by each /n
Worlds_2024_Teams = [Worlds_2024_Teams[i].split('\n') for i in range(len(Worlds_2024_Teams))]
#remove any empty elements from the lists
Worlds_2024_Teams = [[Worlds_2024_Teams[i][j] for j in range(len(Worlds_2024_Teams[i])) if Worlds_2024_Teams[i][j] != ''] for i in range(len(Worlds_2024_Teams))]
#remove the - character from the list
Worlds_2024_Teams = [[Worlds_2024_Teams[i][j].replace('-', '') for j in range(len(Worlds_2024_Teams[i]))] for i in range(len(Worlds_2024_Teams))]

#set a new list called Pokemon which is every 7th element in the list
Pokemon = [[Worlds_2024_Teams[i][j] for j in range(len(Worlds_2024_Teams[i])) if j % 7 == 0] for i in range(len(Worlds_2024_Teams))]
#set a new list called Item which is everything after the @ symbol in Pokemon list
Item = [[Pokemon[i][j].split('@')[1] for j in range(len(Pokemon[i]))] for i in range(len(Pokemon))]
#remove the first character from the list
Item = [[Item[i][j][1:] for j in range(len(Item[i]))] for i in range(len(Item))]
#flatten the list
Item = [item for sublist in Item for item in sublist]

#remove anything after the @ symbol in the Pokemon list
Pokemon = [[Pokemon[i][j].split('@')[0] for j in range(len(Pokemon[i]))] for i in range(len(Pokemon))]

#set a new list called Ability which is every 7th element in the list starting at 1
Ability = [[Worlds_2024_Teams[i][j] for j in range(len(Worlds_2024_Teams[i])) if j % 7 == 1] for i in range(len(Worlds_2024_Teams))]
#Remove 'Ability: ' from the list
Ability = [[Ability[i][j].replace('Ability: ', '') for j in range(len(Ability[i]))] for i in range(len(Ability))]
#flatten the list
Ability = [item for sublist in Ability for item in sublist]

#set a new list called Tera Type which is every 7th element in the list starting at 2
Tera_Type = [[Worlds_2024_Teams[i][j] for j in range(len(Worlds_2024_Teams[i])) if j % 7 == 2] for i in range(len(Worlds_2024_Teams))]
#Remove 'Tera Type: ' from the list
Tera_Type = [[Tera_Type[i][j].replace('Tera Type: ', '') for j in range(len(Tera_Type[i]))] for i in range(len(Tera_Type))]
#flatten the list
Tera_Type = [item for sublist in Tera_Type for item in sublist]

#set a new list called Move_1 which is every 7th element in the list starting at 3
Move_1 = [[Worlds_2024_Teams[i][j] for j in range(len(Worlds_2024_Teams[i])) if j % 7 == 3] for i in range(len(Worlds_2024_Teams))]
#flatten the list
Move_1 = [item for sublist in Move_1 for item in sublist]

#set a new list called Move_2 which is every 7th element in the list starting at 4
Move_2 = [[Worlds_2024_Teams[i][j] for j in range(len(Worlds_2024_Teams[i])) if j % 7 == 4] for i in range(len(Worlds_2024_Teams))]
#flatten the list
Move_2 = [' / '.join(move2) for move2 in Move_2]

#set a new list called Move_3 which is every 7th element in the list starting at 5
Move_3 = [[Worlds_2024_Teams[i][j] for j in range(len(Worlds_2024_Teams[i])) if j % 7 == 5] for i in range(len(Worlds_2024_Teams))]
#flatten the list
Move_3 = [' / '.join(move3) for move3 in Move_3]

#set a new list called Move_4 which is every 7th element in the list starting at 6
Move_4 = [[Worlds_2024_Teams[i][j] for j in range(len(Worlds_2024_Teams[i])) if j % 7 == 6] for i in range(len(Worlds_2024_Teams))]
#flatten the list
Move_4 = [' / '.join(move4) for move4 in Move_4]

#create a new dataframe called Compiled_Teams_2024_Worlds with the columns Pokemon and Item
Compiled_Teams_2024_Worlds = pd.DataFrame(Pokemon, columns = ['Pokemon'])
Compiled_Teams_2024_Worlds['Item'] = Item
Compiled_Teams_2024_Worlds['Ability'] = Ability
Compiled_Teams_2024_Worlds['Tera Type'] = Tera_Type
Compiled_Teams_2024_Worlds['Move 1'] = Move_1
Compiled_Teams_2024_Worlds['Move 2'] = Move_2
Compiled_Teams_2024_Worlds['Move 3'] = Move_3
Compiled_Teams_2024_Worlds['Move 4'] = Move_4
Compiled_Teams_2024_Worlds
```

```python
#add the rank, record, and name columns to the dataframe from the worlds_2024_results dataframe but for each row in the worlds_2024_results dataframe there are 6 rows in the Compiled_Teams_2024_Worlds dataframe
Compiled_Teams_2024_Worlds['Rank'] = [worlds_2024_results['Rank'][i//6] for i in range(len(Compiled_Teams_2024_Worlds))]
Compiled_Teams_2024_Worlds['Record'] = [worlds_2024_results['Record'][i//6] for i in range(len(Compiled_Teams_2024_Worlds))]
Compiled_Teams_2024_Worlds['Name'] = [worlds_2024_results['Name'][i//6] for i in range(len(Compiled_Teams_2024_Worlds))]
#split the Record column into two columns called Wins and Losses
Compiled_Teams_2024_Worlds[['Wins', 'Losses']] = Compiled_Teams_2024_Worlds['Record'].str.split('-', expand = True)
#drop the Record column
Compiled_Teams_2024_Worlds = Compiled_Teams_2024_Worlds.drop(columns = ['Record'])
#add a column titled Tournament with the value of Worlds
Compiled_Teams_2024_Worlds['Tournament'] = 'Worlds'
#drop the name column
Compiled_Teams_2024_Worlds = Compiled_Teams_2024_Worlds.drop(columns = ['Name'])
#save the dataframe to a csv file
#Worlds_2024_Teams_Results = Compiled_Teams_2024_Worlds.to_csv('Worlds_2024_Teams_Results.csv', index = False)
Compiled_Teams_2024_Worlds
```
