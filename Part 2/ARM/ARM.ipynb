```python
import pandas as pd

#loading in the data from the csv file
df = pd.read_csv('C:/Users/ryhen/OneDrive - UCB-O365/Machine Learning/Project/Part 1/Pokemon/Visualizations/Combined_Teams_2024_Updated.csv')
df.head()
```

```python
#removing the columns not useful for ARM 
#removing columns wins, losses, tournament, stat_totla, hp, attack, defense, special_attack, special_defense, speed, move_1_generation, move_1_power,move_1_effect_chance, move_1_accuracy,move_2_generation, move_2_power,move_2_effect_chance, move_2_accuracy, move_3_generation,move_3_generation, move_3_power,move_3_effect_chance, move_3_accuracy,move_4_generation, move_4_power,move_4_effect_chance, move_4_accuracy
df = df.drop(['wins', 'losses', 'tournament', 'stat_total', 'hp', 'attack', 'defense', 'special_attack', 'special_defense', 'speed', 'move_1_generation', 'move_1_power','move_1_effect_chance', 'move_1_accuracy','move_2_generation', 'move_2_power','move_2_effect_chance', 'move_2_accuracy', 'move_3_generation','move_3_power','move_3_effect_chance', 'move_3_accuracy','move_4_generation', 'move_4_power','move_4_effect_chance', 'move_4_accuracy', 'move_1_effect', 'move_2_effect', 'move_3_effect', 'move_4_effect', 'type_one', 'type_two', 'type'], axis=1)
#on the item column replace any spaces with a -
df['item'] = df['item'].str.replace(' ', '-')
df['ability'] = df['ability'].str.replace(' ', '-')
#add 'gen' to the end of each element in the generation column
df['pokemon_generation'] = df['pokemon_generation'].astype(str) + 'gen'
# Remove the column names row (header)
df_no_header = pd.DataFrame(df.values)  # Converts the DataFrame into one without the header
#save the data to a csv file titled ARM_Pokemon.csv
df_no_header.to_csv('ARM_Pokemon.csv', index=False, header=False)
df_no_header
df
```
